```python
monthly_data = {}
monthly_data['month'] = "April"
monthly_data['place_pairs'] = []
#Appending Start and End locations
placePair = {}
placePair["Start"] = "Monroe Park"
placePair["End"] = "Monroe Park"

# appending corresponding route IDs
routeIDs = April_route_report.loc[(April_route_report["Start"]== "Monroe Park") & (April_route_report["End"]== "Monroe Park")]["Route ID"].to_numpy()
placePair["Route IDs"] = routeIDs
monthly_data['place_pairs'].append(placePair)

#appending GPS_locs
gps_points = GPS_data[GPS_data['Route ID'].isin(routeIDs)]
#latlons = gps_points.join(pd.DataFrame())
gps_arrs=[]
for index,points in gps_points.iterrows():
    #print(points["Latitude"])
    gps_arrs.append([points["Latitude"],points["Longitude"],3])
monthly_data['Locations'] = gps_arrs
monthly_data
```

```python
monthly_data = []
month_index=0
for month_data in monthly_reports:
    # each month variable
    month={}
    month['month'] = month_index    
    month_index=month_index+1;
    #Each place pair corresponding to months
    month['place_pairs'] = []
    for k in ["Bike","Pedelec"]:
        for i in start_places:
            for j in end_places:
                # pair init
                pair = {}

                ## adding start and end locations 
                pair["Start"] = i
                pair["End"] = j
                pair['Type'] = k
                
                ## appending corresponding route IDs
                routeIDs = month_data.loc[(month_data["Start"]== i) & (month_data["End"]== j) & (month_data["Type"]==k)]["Route ID"].tolist()
                #pair["Route IDs"] = routeIDs

                ## appending GPS locs
                gps_points = GPS_data[GPS_data['Route ID'].isin(routeIDs)]

                gps_arrs=[]
                for index,points in gps_points.iterrows():
                    #print(points["Latitude"])
                    gps_arrs.append([points["Latitude"],points["Longitude"],3])
                pair['Locations'] = gps_arrs
                

                month["place_pairs"].append(pair)
    print(month_index)
    monthly_data.app end(month)
#monthly_data

```

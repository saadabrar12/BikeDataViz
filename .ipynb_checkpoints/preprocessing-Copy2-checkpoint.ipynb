```python
#columnchart- summing starts
columnchart=df1[['date','Start']].groupby([df1['date'].dt.month,'Start']).agg(outgoing=('Start','count'))
columnchart=columnchart.reset_index()
```

```python
columnchart
```

```python
#columnchart2-summing ends
columnchart2=df1[['date','End']].groupby([df1['date'].dt.month,'End']).agg(incoming=('End','count'))
columnchart2=columnchart2.reset_index()
```

```python
columnchart2
```

```python
#merge incoming and outgoing calculation
mergedcolumnchart=columnchart.merge(columnchart2, how="outer",left_on=["date","Start"], right_on=["date","End"])
print(mergedcolumnchart.isnull().sum())
```

```python
#replace Nan values in Start column with End values
#also replace incoming and outgoing nans with 0
#Dropping End column
mergedcolumnchart.loc[mergedcolumnchart['Start'].isnull(),'Start'] = mergedcolumnchart['End']
mergedcolumnchart['outgoing'].fillna(0, inplace=True)
mergedcolumnchart['incoming'].fillna(0, inplace=True)
mergedcolumnchart.drop('End', axis=1, inplace=True)
print(mergedcolumnchart.isnull().sum())
```
